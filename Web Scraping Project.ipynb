### Extracting Product(Jacket) details from Flipkart

```python
import requests
from bs4 import BeautifulSoup
```

```python
pip install bs4
```

```python
url="https://www.flipkart.com/montrez-full-sleeve-washed-men-jacket/product-reviews/itm52728358a466e?pid=JCKG5HB9D6SEDBGK&lid=LSTJCKG5HB9D6SEDBGKNWYVFA&marketplace=FLIPKART"
```

```python
url
```

```python
page=requests.get(url)
print(page)
```

```python
data=page.text
```

```python
soup=BeautifulSoup(data,'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
title=soup.title
title
```

```python
print(title.text)
```

```python
stars=soup.find_all("div",class_="XQDdHH Ga3i8K _9lBNRY")
cust_stars=[]
for i in stars:
    cust_stars.append(i.text)

cust_stars
```

```python
customers=soup.find_all("p",class_="_2NsDsF AwS1CA MDcJkH")
customers
```

```python
customers_name=[]
for i in customers:
    customers_name.append(i.text)
customers_name
```

```python
comment=soup.find_all("div",class_="_11pzQk")
comment
```

```python
comments=[]
for i in comment:
    comments.append(i.text)
comments
```

```python
import pandas as pd
```

```python
like_dislikes=soup.find_all('span',class_="tl9VpF")
likedislike=[]
for i in like_dislikes:
    likedislike.append(i.text)
print(likedislike)
```

```python
likes=likedislike[::2]
likes
```

```python
dislikes=likedislike[1::2]
dislikes
```

```python
from IPython.display import display, Markdown
```

```python
df=pd.DataFrame()
title="Extracting Product(Jacket) details from Flipkart"

df['Customer Name']=customers_name
df['Comments']=comments
df['Stars']=cust_stars
df['Likes']=likes
df['Dislikes']=dislikes

display(Markdown(f"## {title}"))

display(df)
```
